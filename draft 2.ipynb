```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

print('Training data:', X_train.shape, y_train.shape)
print('Test data:', X_test.shape, y_test.shape)
```

```python
labels= {0: "T-shirt/top", 1: "Trouser", 2: "Pullover", 3: "Dress", 4: "Coat", 5: "Sandal", 6: "Shirt", 7: "Sneaker", 8: "Bag", 9: "Ankle boot"}
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class preparation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self,X, y= None):
        return self
    def transform(self, X):
        X= X.reshape(X.shape[0], X.shape[1] * X.shape[2])/255
        return X
```

```python
X_train1= X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])/255
```

```python
X_test1= X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])/255
```

<h1> XGBoost </h1>

```python
from xgboost import XGBClassifier 
from xgboost import XGBRegressor
```

```python
model = XGBClassifier(preparation(),eval_metric="rmse",verbosity=2, random_state=42) 
```

```python
X_train1= X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])/255
```

```python
X_test1= X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])/255
```

```python
model.fit(X_train1, y_train)
```

```python
xg_predict= model.predict(X_test1)
```

```python
print(accuracy_score(xg_predict, y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
xg_scores = cross_val_score(model, X_train1, y_train, scoring = 'accuracy', cv=10,  n_jobs=-1)
xg_scores
```

<h1> SVC </h1>

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
```

```python
svm_model = make_pipeline(preparation(), SVC(gamma= 'scale', kernel='rbf', C=10, verbose=2))
```

```python
svm_model.fit(X_train, y_train)
```

```python
svm_predict= svm_model.predict(X_test)
```

```python
print(accuracy_score(svm_predict, y_test))
```
